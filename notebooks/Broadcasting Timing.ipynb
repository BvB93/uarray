```python
from uarray import *
from numba import njit
```

```python
def some_function(a, b):
    return (a + b)[10]
```

```python
a = np.arange(1000 * 1000, dtype="float64").reshape(1000, 1000)
b = np.arange(1000, dtype="float64")
```

```python
some_function(a, b)
```

```python
some_function(a, b)
%timeit some_function(a, b)
```

```python
some_fun_njit = njit(some_function)
some_fun_njit(a, b)
```

```python
%timeit some_fun_njit(a, b)
```

```python
opt = optimize(a.shape, b.shape)(some_function)
```

```python
opt(a, b)
```

```python
print(opt.__optimize_steps__['ast_as_source'])
```

```python
opt(a, b)
%timeit opt(a, b)
```

```python
n_opt = njit(opt)
n_opt(a, b)
```

```python
%timeit n_opt(a, b)
```
