```python
import math, numpy as np


def fft(x):
   N = x.shape[0]
   if N == 1:
       return np.array(x)
   e=fft(x[::2])
   o=fft(x[1::2])
   M=N//2
   print(e.shape, o.shape, M)
   l=e + o * math.e**(-2j*math.pi*np.arange(M)/N)
   r=e - o * math.e**(-2j*math.pi*np.arange(M)/N)
   return np.array(l+r)
```
