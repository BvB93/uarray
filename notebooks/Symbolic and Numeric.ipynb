# Symbolic and Numeric

Let's look at a list, where each item is the index times two.

```python
from uarray import *
import operator
```

## Symbolic list

First we create this symbolically, where we don't actually evaluate the list, just create a function for it:

```python
range_list = List.from_abstraction(Abstraction.create(lambda i: i * Natural(2), Natural(None)))
range_list
```

## Numeric list

Now we can create it by storing all of the values:

```python
range_list_concrete = List.create_infer(*(Natural(i * 2) for i in range(5)))
range_list_concrete
```

```python
visualize_progress(range_list_concrete[Natural(3)], clear=True)
```

## Concat

Now let's look at concatting this list with itself, to see how this is different between the two forms:

### Symbolic

```python
visualize_progress(range_list.concat(Natural(5), range_list))
```

### Numeric

```python
visualize_progress(range_list_concrete.concat(Natural(5), range_list_concrete))
```

### Symbolic and Numeric (goes to symbolic)

```python
visualize_progress(range_list_concrete.concat(Natural(5), range_list))
```
