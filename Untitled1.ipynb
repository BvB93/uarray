# `uarray`

```python
from uarray import *
```

## Booleans

```python
true_ = Bool(True)
false_ = Bool(False)
```

```python
t = true_.and_(false_)
t
```

```python
replace(t)
```

```python
type(Abstraction)
```

```python
a = Abstraction.create(lambda a: a.and_(a).equal(a.or_(a)), Bool)
a
```

```python
a(true_)
```

```python
replace(a(true_))
```

## Abstractions

```python
a2 = Abstraction.create_bin(lambda a, b: a.and_(b), Bool, Bool)
a2
```

```python
a2(true_)(true_)
```

## Naturals

```python
n = Nat(10)
```

```python
exp = Abstraction.create(lambda i: (i - Nat(1)).loop(Nat(2), Abstraction.create_bin(lambda acc, i: acc * Nat(2), Nat, Nat)), Nat)

exp
```

```python
replace(exp)
```

```python
replace(exp(Nat(20)))
```

## Lists

```python
range_abstraction = Abstraction.create(
    lambda n: List.create(
        Nat,
        n.loop(
            ListContents.create(),
            Abstraction.create_bin(
                lambda lst, idx: lst.concat(ListContents.create(idx)),
                ListContents,
                Nat
            )
        )
    ),
    Nat
)
```

```python
replace(range_abstraction)
```

```python
range_10 = replace(range_abstraction(Nat(10)))
range_10
```

```python
idxed = range_10[Nat(4)]
idxed
```

```python
replace(range_10[Nat(4)])
```
