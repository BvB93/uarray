# `uarray`

```python
from uarray import *
```

## Booleans

```python
true_ = Bool(True)
false_ = Bool(False)
```

```python
t = true_.and_(false_)
t
```

```python
replace(t)
```

## Abstractions

```python
a2 = Abstraction.create_bin(lambda a, b: a.and_(b), Bool(None), Bool(None))
a2
```

```python
a2(true_)(false_)
```

```python
replace(a2(true_)(true_))
```

## Naturals

```python
n = Nat(10)
```

```python
exp = Abstraction.create(lambda i: (i - Nat(1)).loop(Nat(2), Abstraction.create_bin(lambda acc, i: acc * Nat(2), Nat(None), Nat(None))), Nat(None))

exp
```

```python
replace(exp(Nat(20)))
```

## Lists

```python
l = List.create_infer(Nat(1), Nat(2), Nat(3))
l
```

```python
getitem_abs = Abstraction.create(lambda n: l[n], Nat(None))
getitem_abs
```

```python
replace(getitem_abs(Nat(0)))
```

```python
replace(getitem_abs(Nat(1)))
```

```python
range_abstraction = Abstraction.create(
    lambda n: n.loop(
        List.create(Nat(None)),
        Abstraction.create_bin(
            lambda lst, idx: lst.concat(List.create_infer(idx)),
            List(None, Nat),
            Nat(None),
        )
    ),
    Nat(None)
)
```

```python
range_abstraction
```

```python
range_five = range_abstraction(Nat(5))
replace(range_five)
```

```python
idxed = range_five[Nat(4)]
replace(idxed)
```

## Vectors

```python
v = Vec.create_infer(Nat(1), Nat(3), Nat(5))
v
```

```python
replace(v.concat(v))
```

```python
nat_add_abstraction = Abstraction.create_bin(
    lambda x, y: x + y, Nat(None), Nat(None)
)
nat_add_abstraction
```

```python
reduced = v.reduce(Nat(0), nat_add_abstraction)
```

```python
replace(reduced)
```
