```python
a = Abstraction.create(lambda a: a.and_(a).equal(a.or_(a)), Bool)
a
```

## Lists

```python
l = List.create_infer(Nat(1), Nat(2), Nat(3))
l
```

```python
getitem_abs = Abstraction.create(lambda n: l[n], Nat)
getitem_abs
```

```python
getitem_abs(Nat(0))
```

```python
range_abstraction = Abstraction.create_simple(
    lambda n: n.loop(
        List.create_args(Nat),
        Abstraction.create_bin(
            lambda lst, idx: lst.concat(List.create_infer(idx)),
            ListContents,
            lambda lc: List.create(Nat, lc),
            Nat,
            lambda n: n
        )
    ),
    Nat
)
```

```python
range_abstraction
```

```python
replace(range_abstraction(Nat(5)))
```

```python
range_abstraction = Abstraction.create_simple(
    lambda n: List.create(
        Nat,
        n.loop(
            ListContents.create(),
            Abstraction.create_bin_simple(
                lambda lst, idx: lst.concat(ListContents.create(idx)),
                ListContents,
                Nat
            )
        )
    ),
    Nat
)
```

```python
replace(range_abstraction)
```

```python
range_10 = replace(range_abstraction(Nat(10)))
range_10
```

```python
idxed = range_10[Nat(4)]
idxed
```

```python
replace(range_10[Nat(4)])
```
